```python
on_axis_intensities = np.array([188, 177, 153, 12])
on_axis_errors = np.array([11, 19, 14, 23])/188
on_axis_intensities = on_axis_intensities/np.max(on_axis_intensities)
on_axis_distances = np.array([6, 9, 16, 23])
ka2 = 100*(2*np.pi/(633*(10**(-9))))*((0.5*0.75*(10**(-3)))**(2)) #in cm
print(ka2)
x = np.arange(5, 26.0001, 0.0001)
I = (np.sin(ka2/(2*x)))**2
print(x.shape[0])
#ind1 = 
plt.figure()
matplotlib.rcParams.update({'font.size': 14})
plt.title('Circular Aperture Diffraction: On-Axis Intensity')
plt.xlabel('Distance from filter (cm)')
plt.ylabel('Normalized Intensity')
plt.plot(x, I, 'b', label = 'Theory')
plt.errorbar(on_axis_distances, on_axis_intensities, yerr = on_axis_errors, fmt = 'ro', ms = 8, label = 'Experiment')
plt.legend(bbox_to_anchor = (0.5, 0.3))
plt.tight_layout()
plt.savefig('D:\\Cornell\Cornell Semesters\Semester 2, Fall\'21\PHYS 6510 - Advanced Physics Lab\Experiment - Diffraction\Report\Figures\Circular.jpg')
plt.show()
```
